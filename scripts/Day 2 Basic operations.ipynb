```python
df= (spark.read
.option("header", "true")
.option("inferSchema", "true").csv("/Volumes/workspace/ecommerce/ecommerce_data/2019-Oct.csv"))
```

```python
df.show(10)
```

```python
selected_df= df.select("product_id","category_ids","category_code","brand","price")
```

```python
selected_df.display()
```

```python
df.filter(df.price>1000).show(5)
```

```python
df.filter("price>1000").show(5)
```

```python
from pyspark.sql.functions import count

df.groupBy('brand').agg(count('brand').alias('brand_counts')).display()
```

```python
df.groupBy("category_code").agg(sum('price').alias('total_price')).display()
```

```python
df.select("event_time","event_type","product_id","category_id","category_code","brand","price","user_id","user_session").orderBy("product_id").show(5)
```

```python
df.orderBy("brand").show(5)
```
